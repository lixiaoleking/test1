## 生成新文件

```python
import os
root = r'D:\车联网\data'

# 给定文件夹下面的文件名
filenames = os.listdir(root)

for name in filenames:
    filepath = root+'\\'+name
    #print(filepath)
    detail = pd.read_excel(filepath,sep=',', encoding='utf-8')
    print(str(name)+"的文件行列为：",detail.shape)
    for i in detail.columns:
        if detail[i].count() == 0:
            detail.drop(labels=i, axis=1, inplace=True)
    detail.to_excel(root+'\\new_'+name+'.xlsx')
    print("修改后的"+str(name)+"的文件行列为：",detail.shape)
    print()
```

```python
import os
def new_file(path,to_path):    
    root = path
    # 给定文件夹下面的文件名
    filenames = os.listdir(root)

    for name in filenames:
        filepath = root+'\\'+name
        #print(filepath)
        detail = pd.read_excel(filepath,sep=',', encoding='utf-8')
        print(str(name)+"的文件行列为：",detail.shape)
        for i in detail.columns:
            if detail[i].count() == 0:
                detail.drop(labels=i, axis=1, inplace=True)
        detail.to_excel(to_path+'\\new_'+name)
        print("修改后的"+str(name)+"的文件行列为：",detail.shape)
        print()
new_file(r'D:\车联网\data',r'D:\车联网\step1')
```
